```python
new_books = books.drop_duplicates("Book-Title")
new_books.info()
```

```python
ratings_with_name = ratings.merge(new_books, on="ISBN")
ratings_with_name.drop(["ISBN", "Image-URL-S", "Image-URL-M", "Image-URL-L", "Publisher"], axis=1, inplace=True)
ratings_with_name.head()
```

```python
users_ratings_matrix = ratings_with_name.merge(users, on="User-ID")

print(len(ratings_with_name["User-ID"].unique()))
```

```python
ratings_with_name.dropna(inplace=True)
ratings_with_name.info()
```

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 100
expert_users = x[x].index

filtered_user_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(expert_users)]

y = filtered_user_ratings.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = y[y].index
user_ratings = filtered_user_ratings[filtered_user_ratings["Book-Title"].isin(famous_books)]
```

```python
design_matrix = user_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")

design_matrix.fillna(0, inplace=True)
design_matrix.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

scaler = StandardScaler(with_mean=True, with_std=True)
design_matrix_centered = scaler.fit_transform(design_matrix)
design_matrix_centered
```

```python
similiarity_score = cosine_similarity(design_matrix_centered)
similiarity_score
```

```python
import numpy as np

def recommend(book_name):
    index = np.where(design_matrix.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similiarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    data = []

    for index, similarity in similar_books:
        item = []
        temp_df = new_books[new_books["Book-Title"]==design_matrix.index[index]]
        item.extend(temp_df["Book-Title"].values)
        item.extend(temp_df["Book-Author"].values)
        data.append(item)
    return data
```

```python
recommend('The Hot Zone')
```

```python
from sklearn.neighbors import KNeighborsTransformer
k = 5
model = KNeighborsTransformer(n_neighbors=k, n_jobs=-1, mode="distance", metric="cosine")
transform = model.fit_transform(design_matrix_centered)

book_title = 'Solyaris'
book_index = np.where(design_matrix.index==book_title)[0][0]
print(book_index)
csr = transform[book_index]
print(csr)
#print(design_matrix.iloc[i].name)
```
