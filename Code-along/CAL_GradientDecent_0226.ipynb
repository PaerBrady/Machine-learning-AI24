```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

plt.style.use("dark_background")
samples = 1000

X = 2*np.random.rand(samples, 1)
y = 2+9*X+np.random.randn(samples,1)

plt.plot(X,y, '.');
```

```python
X = np.c_[np.ones(samples), X]
X[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# adds up to 1000
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def gradient_decent(X, y, learning_rate = .1, iterations = 100):
    m = len(X)

    theta = np.random.rand(X.shape[1], 1)
    
    for i in range(iterations):
        gradient = (2/m) * X.T @ ((X @ theta) - y)
        theta = learning_rate * gradient

    return theta

theta = gradient_decent(X_train, y_train)
theta
```

```python
steps = range (1, 200, 1)
thetas = [gradient_decent(X_train, y_train, iterations=iteration) for iteration in steps]
thetas = np.array(thetas)
thetas.shape
```
